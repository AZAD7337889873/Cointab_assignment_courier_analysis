```python
# calculating total order price
order["total order price in Rs"] = order["Order Qty"]*order["Item Price(Per Qty.)"]
order.head()
```

```python
# COD charges

charge = []
for x in order["Payment Mode"]:
  if x == "COD":
    for y in order["total order price in Rs"]:
      if y<= 300:
        charge.append(15)
      else:
        charge.append(y*0.05)
  else:
    charge.append(0)
charge[0:10]
```

```python
order["COD charges in Rs as per X"] = charge[0:400]
order["COD charges in Rs as per X"] = np.where(order["Payment Mode"]=="Prepaid",0,order["COD charges in Rs as per X"])
```

```python
# weight slab for zone_b as per X
def weight_slab(weight):
  i = round(weight%1,1)
  if i == 0.0:
    return weight
  elif i > 0.5:
    return int(weight)+1
  else:
    return int(weight)+0.5

zonexb["weight slab as per X (kg)"]= zonexb["total_weight in KG as per X"].apply(weight_slab)
zonexb
```

```python
# weight slab for zone_d as per X
def weight_slab(weight):
  i = round(weight%1.25,2)
  if i == 0.00:
    return weight
  elif weight<1.25:
    return 1.25
  else:
    return (weight//1.25)*1.25+1.25

zonexd["weight slab as per X (kg)"]= zonexd["total_weight in KG as per X"].apply(weight_slab)
zonexd.head()
```

```python
# weight slab for zone_e as per X
def weight_slab(weight):
  i = round(weight%1.5,2)
  if i == 0.00:
    return weight
  elif i <1.5:
    return 1.5
  else:
    return (weight//1.5)*1.5+1.5

zonexe["weight slab as per X (kg)"]= zonexe["total_weight in KG as per X"].apply(weight_slab)
zonexe.head()
```

```python
#Expected charge for X in b zone
fwd_fxd = 33
fwd_add = 28.3
rto_fxd = 20.5
rto_add = 28.3
weight_slab = zonexb["weight slab as per X (kg)"]
add_weight = (weight_slab-0.5)/0.5
zonexb["expected charges by X in Rs"] = 0
zonexb["expected charges by X in Rs"] = np.where(zonexb['Type of Shipment'] == 'Forward charges',(fwd_fxd + add_weight*fwd_add), zonexb["expected charges by X in Rs"])
zonexb["expected charges by X in Rs"] = np.where(zonexb["Type of Shipment"]== "Forward and RTO charges",(fwd_fxd + rto_fxd + add_weight*(fwd_add + rto_add)) ,zonexb["expected charges by X in Rs"])
zonexb.head()
```

```python
#Expected charge for X in d zone
fwd_fxd = 45.4
fwd_add = 44.8
rto_fxd = 41.3
rto_add = 44.8
weight_slab = zonexd["weight slab as per X (kg)"]
add_weight = (weight_slab-1.25)/1.25
zonexd["expected charges by X in Rs"] = 0
zonexd["expected charges by X in Rs"] = np.where(zonexd['Type of Shipment'] == 'Forward charges',(fwd_fxd + add_weight*fwd_add), zonexd["expected charges by X in Rs"])
zonexd["expected charges by X in Rs"] = np.where(zonexd["Type of Shipment"]== "Forward and RTO charges",(fwd_fxd + rto_fxd + add_weight*(fwd_add + rto_add)) ,zonexd["expected charges by X in Rs"])
zonexd.head()
```

```python
#Expected charge for X in e zone
fwd_fxd = 56.6
fwd_add = 55.5
rto_fxd = 50.7
rto_add = 55.50
weight_slab = zonexe["weight slab as per X (kg)"]
add_weight = (weight_slab-1.5)/1.5
zonexe["expected charges by X in Rs"] = 0
zonexe["expected charges by X in Rs"] = np.where(zonexe['Type of Shipment'] == 'Forward charges', (fwd_fxd + add_weight*fwd_add), zonexe["expected charges by X in Rs"])
zonexe["expected charges by X in Rs"] = np.where(zonexe["Type of Shipment"]== "Forward and RTO charges",(fwd_fxd + rto_fxd + add_weight*(fwd_add + rto_add)) ,zonexe["expected charges by X in Rs"])
zonexe.head()
```

## calculations of weight slab and total price


```python
# weight slab for zone_b as per Courier
def weight_slab(weight):
    i = round(weight % 1, 1)
    if i == 0.0:
        return weight
    elif i > 0.5:
        return int(weight)+1
    else:
        return int(weight)+0.5

c_b["Weight slab charged by Courier in kg"] = c_b["total weight as per courier"].apply(weight_slab)
c_b.head()
```

```python
# weight slab for zone_d as per courier
def weight_slab(weight):
    i = round(weight % 1.25, 2)
    if i == 0.00:
        return weight
    elif weight<1.25:
        return 1.25
    else:
        return (weight//1.25)*1.25+1.25

c_d["Weight slab charged by Courier in kg"] = c_d["total weight as per courier"].apply(weight_slab)
c_d.head()
```

```python
# weight slab for zone_e as per courier
def weight_slab(weight):
    i = round(weight % 1.5, 2)
    if i == 0.00:
        return weight
    elif weight<1.5:
        return 1.5
    else:
        return (weight//1.5)*1.5+1.5

c_e["Weight slab charged by Courier in kg"] = c_e["total weight as per courier"].apply(weight_slab)
c_e.head()
```

```python
data_done = pd.concat([c_b,c_d,c_e])
data_done.head()
```

```python
data_done["Expected Charges as per X in Rs"] = data_done["expected charges by X in Rs"]+data_done["COD charges in Rs as per X"]
data_done.head()
```

```python
data_done.rename(columns= {"Billing Amount (Rs.)":"Charges Billed by Courier Company in Rs"}, inplace = True)
data_done["Difference Between Expected Charges and Billed Charges (Rs.)"] = data_done["Charges Billed by Courier Company in Rs"] - data_done["Expected Charges as per X in Rs"]
data_done
```



> preparing worksheet 1



```python
# Calculate the total orders in each category
total_correctly_charged = len(output_file[output_file['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0])
total_overcharged = len(output_file[output_file['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0])
total_undercharged = len(output_file[output_file['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0])

# Calculate the total amount in each category
amount_overcharged = abs(output_file[output_file['Difference Between Expected Charges and Billed Charges (Rs.)']> 0]['Difference Between Expected Charges and Billed Charges (Rs.)'].sum())
amount_undercharged = output_file[output_file['Difference Between Expected Charges and Billed Charges (Rs.)']< 0]['Difference Between Expected Charges and Billed Charges (Rs.)'].sum()
amount_correctly_charged = output_file[output_file['Difference Between Expected Charges and Billed Charges (Rs.)']== 0]['Expected Charge as per X (Rs.)'].sum()

# Create a new DataFrame for the summary
summary_data = {'Description': ['Total Orders where X has been correctly charged',
                                'Total Orders where X has been overcharged',
                                'Total Orders where X has been undercharged'],
                'Count': [total_correctly_charged, total_overcharged, total_undercharged],
                'Amount (Rs.)': [amount_correctly_charged, amount_overcharged, amount_undercharged]}

df_summary = pd.DataFrame(summary_data)

print(df_summary)
```
